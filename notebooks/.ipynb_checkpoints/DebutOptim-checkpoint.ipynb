```python
import numpy as np 
import sympy as sp
from matplotlib import pyplot as plt
```

## Dérivée

Travaillons avec une fonction $f=(\alpha -\frac{\beta}{2}x)x$. 

```python
def f(x,alpha,beta):
    return (alpha - 0.5*beta*x) * x
```

```python
x = np.linspace(0,20,10)
alpha = 10
beta = 1.0
y = f(x,alpha,beta)
```

```python
plt.figure()
plt.plot(x,y)
plt.show()
```

On peut utiliser SimPy: 

```python
x,a,b = sp.symbols('x,a,b')
```

```python
sf = (a-b/2*x)*x
sf
```

```python
sp.plot(sf.subs(a,10).subs(b,1),('x',0,20))
```

On peut voir comment la fonction se comporte quand on change $\alpha$:

```python
alphas = np.linspace(5,15,10)
beta = 1.0
plt.figure()
for alpha in alphas: 
    plt.plot(x,f(x,alpha,beta),label=str(np.round(alpha,2)))
plt.legend(loc=2)
plt.title("Changer $\\alpha$")
plt.show()
```

Même chose avec beta: 

```python
betas = np.linspace(0.5,1.5,10)
alpha = 10.0
plt.figure()
for beta in betas: 
    plt.plot(x,f(x,alpha,beta),label=str(np.round(beta,2)))
plt.legend(loc=2)
plt.title("Changer $\\beta$")
plt.show()
```

Maintenant programmons la dérivée de la fonction. La dérivée est $$ f'(x) = \alpha - \beta x $$ 

```python
def df(x,alpha,beta):
    return alpha - beta*x
```

Graphiquement on a :

```python
alpha = 10
beta = 1.0
y = f(x,alpha,beta)
dy = df(x,alpha,beta)
plt.figure()
plt.plot(x,y,label='fonction')
plt.plot(x,dy,label='dérivée')
plt.axvline(x=10.0,linestyle=':')
plt.plot(x,np.zeros(len(x)),linestyle=':')
```

On peut trouver aussi la dérivée avec SimPy:

```python
dsf = sp.diff(sf,x)
dsf
```

```python
sp.plot(dsf.subs(a,10).subs(b,1),('x',0,20))
```

On peut écrire une fonction à partir de notre dérivée SimPy

```python
dfs = sp.lambdify((x,a,b),dsf)
```

```python
dfs(0,10,1),dfs(5,10,1)
```

Peut-on trouver une dérivée numériquement (quand on ne connait pas la fonction)? Oui, on fait des différences petites: 

```python
def dfn(x,alpha,beta,eps=1e-6):
    return (f(x+eps,alpha,beta) - f(x-eps,alpha,beta))/(2*eps)
```

```python
alpha = 10
beta = 1.0
dy = df(x,alpha,beta)
dny = dfn(x,alpha,beta)
plt.figure()
plt.scatter(x,dny,label='numérique')
plt.plot(x,dy,label='exacte')
```

Pas mal comme approximation...

## Maximisation sans contrainte

Reprenons la même fonction. Nous cherchons le maximum. Bien sur il est donnée par :
    $$ f'(x) = \alpha - \beta x^* = 0 $$
et donc $x^* = \frac{\alpha}{\beta}$. 

```python
xopt = alpha/beta
xopt
```

On appelle cette solution la solution analytique. Bien sur, on peut la trouver numériquement. Commencons par regarder sur une grille la valeur de la dérivée: 

```python
dy = df(x,alpha,beta)
[(i,di) for i,di in zip(x,dy)]
```

On voit que entre 8.88 et 11.11, la derivée change de signe. Donc, on pourrait chercher dans cet interval avec une nouvelle grille

```python
xx = np.linspace(8.88,11.11,10)
[(i,di) for i,di in zip(xx,df(xx,alpha,beta))]
```

On voit qu'il y a changement de signe entre 9.87 et 10.11. On peut donc créer une nouvelle grille jusqu'à ce que la borne inférieure et supérieur soit très près de l'autre (disons 1e-6). C'est une façon de faire mais ce n'est pas très efficace. Une autre façon est d'utiliser ce qu'on appelle la bissection. La bissection est très simple (voir [cet article](https://fr.wikipedia.org/wiki/M%C3%A9thode_de_dichotomie) wikipedia si cela vous intéresse). Elle regarde un point entre le maximum et le minimum. Si la dérivé a le même signe que le maximum, on sait que le minimum se trouve en dessous de ce maximum. Donc, on remplace le maximum par ce point. On fait la même chose de l'autre côté.
