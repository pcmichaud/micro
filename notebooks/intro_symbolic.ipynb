```python
import sympy as sp
```

```python
x, a, b = symbols('x a b')
```

```python
y = a*x**b
y
```

```python
fy = y.subs(a,1.0).subs(b,0.5)
fy
```

```python
sp.plot(fy,('x',0,10))
```

```python
def gy(x,a,b):
    return a*x**b
```

```python
gy(1.0,1.0,0.5)
```

```python
import numpy as np 
```

```python
xs = [x for x in np.linspace(0,10,100)]
```

```python
ys = [gy(x,1.0,0.5) for x in xs]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(xs,ys)
```

```python
approx = fy + diff(fy)
approx
```

```python
x0 = symbols('x0')
```

```python
approx = fy.subs(x,x0) + diff(fy).subs(x,x0) * (x - x0)
approx
```

```python
p0 = sp.plot(fy,('x',0,10),show=False) 
p1 = sp.plot(approx.subs(x0,3),('x',1,7),show=False)
p0.append(p1[0])
p0.show()
```
